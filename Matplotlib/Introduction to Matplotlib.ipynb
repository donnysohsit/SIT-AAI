## Section 6a: Introduction to MatplotLib

Although matplotlib offers several ways to plot graphs, typically only two methods are used. They are plot and scatter. This section will go through some of the syntax of these two plotting methods. We will then follow up with how the plotting methods are used in a typical data science project, both for clustering (Section 6b) and a bit of classification (Section 6c)

### Section 6a.1: Plot

This creates 100 spaced values values from 0 to 10.

```
x = np.linspace(0, 10, 100)
```


We can set the way the lines are shown with the next parameter. here we can choose a dash line:



```
plt.plot(x, np.cos(x), '--');
```
